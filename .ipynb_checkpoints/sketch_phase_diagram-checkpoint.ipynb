```julia
using DelimitedFiles, Plots, Statistics

using CurveFit, LaTeXStrings

Plots.default(
    grid = true,
    size = (400,400),
    fontfamily="serif-roman", 
    labelfontsize=15,
    titlefontsize =12,
    xguidefontsize=15,
    yguidefontsize=15,
    xtickfontsize=12,
    ytickfontsize=12,
    markerstrokewidth = 1,        # border around dots
#    markersize = 3
)
```

```julia
x0 = 0.25:0.001:0.75
x1 = ones(length(x0)).*0.5;
```

```julia
pdiag = plot(grid=false,axis=([], false), size = (600,100), xlims=(0,1.1), ylims=(-0.15,0.5), margin=4*Plots.mm)

plot!(pdiag, x0, x1, fillrange = zero(x0), fc=:blues, label="", color=:blues)

plot!(pdiag, [0,1],[0,0],arrow=(:closed, 2.0), color="black", label="")

plot!(pdiag, [0.25,0.25], [-0.1, 0.1], label="", color="black")

plot!(pdiag, [0.75,0.75], [-0.1, 0.1], label="", color="black")

plot!(pdiag, [0.8,0.8], [-0.2,0.2], color=1, label="")
plot!(pdiag, [0.7,0.7], [-0.2,0.2], color=2, label="")

annotate!(pdiag, 0.5,0.25, "RSB")
annotate!(pdiag, 0.25,-0.25, L"\gamma_{\rm stab}^1")
annotate!(pdiag, 0.75,-0.25, L"\gamma_{\rm stab}^2")
annotate!(pdiag, 1.02,-0.1, L"\gamma")
```

```julia
savefig(pdiag, "phase_diagram.pdf")
```

```julia
x0BP = 0.35:0.001:0.65
x1BP = ones(length(x0BP)).*0.6;
```

```julia
pdiag = plot(grid=false,axis=([], false), size = (600,100), xlims=(0,1.1), ylims=(-0.23,0.5), margin=4*Plots.mm)

plot!(pdiag, x0, x1, fillrange = zero(x0), fc=:blues, label="", color=:blues)

plot!(pdiag, x0BP, x1BP, fillrange = zero(x0BP), fc=:reds, label="", color=:blues)

plot!(pdiag, [0,1],[0,0],arrow=(:closed, 2.0), color="black", label="")

plot!(pdiag, [0.25,0.25], [-0.1, 0.1], label="", color="black")

plot!(pdiag, [0.75,0.75], [-0.1, 0.1], label="", color="black")


annotate!(pdiag, 0.5,0.25, "BP unstable")
annotate!(pdiag, 0.25,-0.25, L"\gamma_{\rm stab}^1")
annotate!(pdiag, 0.75,-0.25, L"\gamma_{\rm stab}^2")
annotate!(pdiag, 1.02,-0.1, L"\gamma")

plot!(pdiag, [0.35,0.35], [-0.1, 0.1], label="", color="black")
plot!(pdiag, [0.65,0.65], [-0.1, 0.1], label="", color="black")
annotate!(pdiag, 0.35,-0.3, L"\gamma_{\rm stab}^{BP,1}")
annotate!(pdiag, 0.65,-0.3, L"\gamma_{\rm stab}^{BP,2}")
```

```julia
savefig(pdiag, "phase_diagram_BP.pdf")
```
